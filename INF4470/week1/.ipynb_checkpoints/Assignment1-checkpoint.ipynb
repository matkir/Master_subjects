### b) 
*Vis at $(cos(θ) + jsin(θ))^n = (cos(θn) + jsin(θn))$ *


Vi definerer $e^{j\theta} = (cos(\theta)+j sin(\theta))$

Dermed kan vi ta
$$ e^{(j\theta)^{n}}  =>  (cos(\theta)+j sin(\theta))^n $$

Vi kan nå bruke potensoperasjoner for å gjøre:

$ e^{(j\theta n)}$ som samsvarer tilsier at: 

$$ e^{(j\theta n)} = (cos(\theta n)+j sin(\theta n)) $$

altså:

$$ (cos(\theta )+j sin(\theta ))^n= (cos(\theta n)+j sin(\theta n)) $$




## Oppgave 4 Komplekse tall og det komplekse tallplanet


```python
from math import atan2,sqrt
# z = a + jb
a=0
b=1
r=sqrt(a**2+b**2)
theta=atan2(b,a)

kart=complex(a,b)
polar=(r,theta)

print('kartesisk = {0}'.format(kart))
print('polar = {0}'.format(polar))
```

### b)
Ved å gange med den komplekskonjugerte vil vi fjerne alle imaginære ledd fra likningen:

Dette er fordi $j^2=-1$

```python
from numpy import *
import matplotlib.pyplot as plt

a=2
b=1

num1=complex(a,b)
num2=complex(a,-b)
num3=num1*num2
print(num1,'*',num2,'=',num3)

plt.plot([real(num1)],[imag(num1)],'bo')
plt.plot([real(num2)],[imag(num2)],'ro')
plt.plot([real(num3)],[imag(num3)],'go')
plt.xlim(0,6)

plt.show()
```

![euelers](img2.png)
Her ser vi at $\theta$ (eller *x* i figur) er vinkel som spenner seg ut i det imaginære planet
og ettersom $\theta$ øker, vil vi bevege oss rundt planet, helt til $\theta=2\pi$, hvor vi har gått en hel runde.

Å opphøye med k, er det samme som å gange $\theta$ med k

```python
from numpy import *
import matplotlib.pyplot as plt

k=array([1,2,3,4,5])
theta=pi/4
tmp1=cos(theta*k)
tmp2=sin(theta*k)


plt.plot(tmp1,tmp2,'ro')

t=linspace(0,2*pi,10000)
plt.plot(cos(t),sin(t))

plt.axis('equal')
plt.show()
```

Velger vi $\theta = 2\pi$ vil vi ha $e^{j\theta k}=1$ uavhengig av k
