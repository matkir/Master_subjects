## Oppgave 4 Komplekse tall og det komplekse tallplanet


```python
from math import atan2,sqrt
# z = a + jb
a=0
b=1
r=sqrt(a**2+b**2)
theta=atan2(b,a)

kart=complex(a,b)
polar=(r,theta)

print('kartesisk = {0}'.format(kart))
print('polar = {0}'.format(polar))
```

### b)
Ved å gange med den komplekskonjugerte vil vi fjerne alle imaginære ledd fra likningen:

Dette er fordi $j^2=-1$

```python
from numpy import *
import matplotlib.pyplot as plt

a=2
b=1

num1=complex(a,b)
num2=complex(a,-b)
num3=num1*num2
print(num1,'*',num2,'=',num3)

plt.plot([real(num1)],[imag(num1)],'bo')
plt.plot([real(num2)],[imag(num2)],'ro')
plt.plot([real(num3)],[imag(num3)],'go')
plt.xlim(0,6)

plt.show()
```
