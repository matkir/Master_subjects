```python
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft,ifft
import numpy as np
rate, data = wav.read('baddolly .wav')

fft_out = np.fft.fft2(data)

plt.title("before")
plt.plot(np.abs(fft_out))
plt.show()

#removing the reqs
for a,i in enumerate(fft_out):
    for j in i:
        if abs(j) > 1E8:
            fft_out[a]=0
        

plt.title("after")
plt.plot(np.abs(fft_out))
plt.show()
new=np.int16(np.fft.ifft2(fft_out));

wav.write('test.wav', rate, new)
```

```python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 1000);
z = np.exp(1j*omega);
h = (2*z**2+3.125)/(z**2-0.9*z + 0.81);
hmin = (2*z**(-2)+3.125)/(1-0.9*z**(-1) + 0.81*z**(-2));


plt.subplot(221)
plt.plot(omega, abs(h))
plt.subplot(223)
plt.plot(omega, abs(hmin))


plt.subplot(222)
plt.plot(omega, np.angle(h))
plt.subplot(224)
plt.plot(omega, np.angle(hmin))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

omega = np.linspace(-np.pi, np.pi, 1000);
z = np.exp(1j*omega);
h = (z*np.pi-z**2)/((1-0.95*np.exp(1j*np.pi/4)/z)*((1-0.95*np.exp(-1j*np.pi/4)/z)))


plt.subplot(211)
plt.plot(omega, abs(h)/max(abs(h)))

plt.subplot(212)
plt.plot(omega, np.angle(h))

plt.show()
```
